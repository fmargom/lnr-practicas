```python
#Some figures
nwords = 0 # number of words
voc, dtags = set(), set() # vocabulary, different tags

for s in brown.tagged_sents():
    #print(s)
    ws = [w for w, _ in s]
    ts = [t for _, t in s]
    nwords += len(ws)
    voc.update(ws)
    dtags.update(ts)

print(f"Number of words: {nwords}")
print(f"Vocabulary size: {len(voc)}")
print(f"Number of different tags: {len(dtags)}")
print("First sentence tagged:")
brown.tagged_sents()[0]
```

## **Tokenizing sentences and words with NLTK**

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt_tab')
# Text from "https://www.nytimes.com/2024/02/20/world/europe/interpol-strongmen-abuse.html"

text="""For years, strongmen and autocrats had a novel weapon in their hunt for political enemies.
They used Interpol, the world’s largest police organization, to reach across borders and grab them — even in democracies.
An award-winning Venezuelan journalist was detained in Peru. An Egyptian asylum seeker was stopped in Australia.
And Russia has tried repeatedly to secure the arrest of William F. Browder, a London-based human rights campaigner."""

sents = sent_tokenize(text)
sents
```

```python
sent_tokens=[]
for s in sents:
    aux = word_tokenize(s)
    sent_tokens.append(aux)
    print(aux)
print(sent_tokens)
```

## **Removing the stopwords in sentences**

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stopw = set(stopwords.words("english"))


for sent in sent_tokens:
    print([x for x in sent if x.lower() not in stopw])
```

## **Determining the Part of the Speech (PoS) of words in a sentence**

For example:

"The young boy can open the can with his hands."
"The/DT young/ADJ boy/NN can/AUX open/VERB the/DT can/NN with/PREP his/PRON hands/NN" ./PUNT

You can use the tagger included in NLTK, e.g. averaged_perceptron_tagger_eng


```python
nltk.download('averaged_perceptron_tagger_eng')
s = "The young boy can open the can with his hands."
untagged_text = word_tokenize(s)
print(untagged_text)
#The input is a list of tokens. The output is a list of tuples (token,tag)
print(nltk.pos_tag(untagged_text)) # This method is used to label a words sequence.
```

## **Training you own HMM-based POS tagger**

```python
import nltk.tag as tagger
from nltk.tag import hmm
from nltk.corpus import brown
from sklearn.model_selection import train_test_split

# Split the corpus in train (90%) and test (10%) partitions
# Here, we have selected only the fist 3000 sentences of the corpus Brown for training.
# Training the entire corpus in Google Colab takes about 8 minutes.

corpus = brown.tagged_sents()[0:3000]
train, test = train_test_split(corpus,random_state=1234, test_size=0.1,shuffle=True)

# Train the model
mytagger = tagger.HiddenMarkovModelTagger.train(train)

# Evaluate the model
mytagger.accuracy(test) #This method evaluates a tagged-sequence. How the predicted labels match with the ground truth labels in the corpus

# Tag a new text

print(mytagger.tag(untagged_text)) # This method is used to label a words sequence.

```

```python
print(mytagger.confusion(test))#This method returns the confusion matrix
```

## **Splitting a corpus in K-FOLD partitions using SKLEARN**

```python
from sklearn.model_selection import KFold
import numpy as np

skf = KFold(n_splits=10)
#filter_corpus=[]
print("Starting a cross-validation schema")
avg_acc=[]
for i, (train_index, test_index) in enumerate(skf.split(corpus)):
    print(f"Iteration {i+1}/{10}")
    train = [corpus[j] for j in train_index]
    test = [corpus[j] for j in test_index]
    #Here you have to train the model using the statement below
    tagger_hmm=hmm.HiddenMarkovModelTagger.train(train)
    #Here you can validate the model using the statement below
    acc=tagger_hmm.accuracy(test)
    print("Partition:",i, "acc:",acc)
    avg_acc.append(acc)

print("Mean of Acc:", np.mean(avg_acc))
```
