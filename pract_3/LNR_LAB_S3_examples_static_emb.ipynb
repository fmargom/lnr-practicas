###EXAMPLE 1.8: Learning word embeddings vector on a specific corpus

```python
import nltk
nltk.download("punkt_tab")
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
MyCorpus = ["Yo it shirt he gave New York was funny.",
"my husband is sick, homemade chicken soup loading husband.",
"Witch raises wind to break up enemy’s lumber pound.",
"i got a new shirt at work at it is the WORST material ever",
"wave is so poor the girls can’t even dress up for Halloween",
"The theory of paint indicates feelings of isolation in society.",
"Wish sometimes I had access to the ever elusive cock carousel..",
"off probation tomorrow ima be a free woman again, FUCK THE SYSTEM"   ]

# Tokenize the corpus. Notice that the preprocessing and tokenization of the texts can be customized depending on the task.

tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in MyCorpus]
# Train the Word2Vec model
model = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=5, min_count=1, workers=4)
# Get the word vector for a specific word
word_vector = model.wv["husband"]
# Most similar words to "example"
similar_words = model.wv.most_similar("husband")
print("Word vector for 'husband':", word_vector)
print("Most similar words to 'husband':", similar_words)
model.wv.save_word2vec_format("my-word-embedding-vectors-100d.txt", binary=False)
```

###EXAMPLE 1.9: Obtaining the representation of a text using word embedding vectors

```python
#Obtaining the words or terms in the text. This time we considered a simple whitespace splitting.
#Notice that, after split the texts, the stopwords can be removed
#dim is the dimension of the word embedding vectors.
def get_sentence_rep(text, keyedvec, dim=300):
    words= text.lower().split()
    zero_vec=np.zeros(dim)
    avg_vec=np.zeros(dim)
    total_w=0
    for w in words:
        try:
            avg_vec+=keyedvec[w]
            total_w+=1
        except:
           print(f"The word {w} is not found in the embedding matrix")
    if total_w == 0:
        return zero_vec
    return avg_vec/total_w

text= "The theory of paint indicates feelings of isolation in society"
get_sentence_rep(text, wordv, dim=200)
```
