# Show Results

```python
from sklearn.metrics import f1_score, classification_report, multilabel_confusion_matrix

def show_result_binary(ref, pred, classes=None):
    # Computing the F1 Score for the positive class (sexism), subtask 2.1
    f1_positive = f1_score(ref, pred, zero_division=0)
    # Obtainig a detailed classification report
    report = classification_report(ref, pred, digits=4, zero_division=0, labels=[0, 1], target_names=classes if classes is not None else ["0", "1"])
    print(f"F1-score (Positive Class): {f1_positive}")
    print(report)

def show_result_multilabel(ref, pred):
    # Computing the macro-average F1 for the second subtask
    f1_macro = f1_score(ref, pred, average='macro', zero_division=0)
    # Obtainig a detailed classification report
    report = classification_report(ref, pred, digits=4, zero_division=0, labels=[0, 1, 2, 3, 4],target_names=LABELS_23)
    print(f"F1-score (Macro-Averaged): {f1_macro}")
    print(report)
    mcm = multilabel_confusion_matrix(ref, pred)
    for i, matrix in enumerate(mcm):
        name = LABELS_23[i].ljust(30)
        print(f"{name}:   TN: {matrix[0,0]:<5} | FP: {matrix[0,1]}  |  FN: {matrix[1,0]:<5} | TP: {matrix[1,1]}")
```
