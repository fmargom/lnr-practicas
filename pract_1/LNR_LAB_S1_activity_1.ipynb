# Create tokenize function

```python
import re
import emoji

# 0 — URLs, e-mails, times (before words and numbers)
url = r"https?://[^\s]+"
email = r"[\w.-]+@[\w.-]+\.[A-Za-z]{2,}"
time = r"\d{1,2}:\d{2}(?:\s*[ap]\.m\.)?"

# 1 — Dates: full (with year) before short dd/mm and dd-mm fragments
date_full_dash = r"\d{1,2}-\d{1,2}-\d{4}"
date_full_slash = r"\d{1,2}/\d{1,2}/\d{2,4}"
date_slash = r"\d{1,2}/\d{1,2}"
date_dash_short = r"\d{1,2}-\d{1,2}"

# 2 — Ellipsis and decimals (comma or dot; optional %)
ellipsis = r"\.\.\."
decimal = r"\d+[.,]\d+%?"

# 3 — Acronyms: CC.OO.; U.R.S.S./S.L./U.S.A.; D.; USA-style caps
cc_oo = r"[A-Z]{2}\.[A-Z]{2}\."
dotted_acronym = r"[A-Z](?:\.[A-Z])+\."
single_letter_acronym = r"[A-Z]\."
caps_word = r"[A-Z]{2,}"

# 4 — Hashtags and @mentions
hashtag = r"#[\w-]+"
mention = r"@[\w_]+"

# 5 — Words with hyphens, then plain words, then integers
hyphen_word = r"[a-zA-Z]+(?:-[a-zA-Z]+)+"
word = r"[a-zA-Z]+"
integer = r"\d+"

# 6 — Punctuation as separate tokens
punct = r'[().,;:%¿?¡!\"\']'

# 7
# 8 — Emoticons / emoji sequences (longest matches first)
emojis = sorted(emoji.EMOJI_DATA, key=len, reverse=True)
pattern_emoji = "(" + "|".join(re.escape(u) for u in emojis) + ")"

pattern_all = "|".join(
    [
        url,
        email,
        time,
        date_full_dash,
        date_full_slash,
        date_slash,
        date_dash_short,
        ellipsis,
        decimal,
        cc_oo,
        dotted_acronym,
        single_letter_acronym,
        caps_word,
        hashtag,
        mention,
        pattern_emoji,
        hyphen_word,
        word,
        integer,
        punct,
    ]
)

re_all = re.compile(pattern_all, re.U)


def tokenize(t, reg=re_all):
    tokens = [t[match.start() : match.end()] for match in reg.finditer(t)]
    return tokens
```

# Test tokenizer
    

```python
import json

def test_tokenizer(tk_function, reference_file):
    with open(reference_file, encoding="utf8") as fh:
        js = json.load(fh)
    N, oks, errors = len(js), 0, 0
    for i, dic in enumerate(js):
        print(f"Sentence {i:02} ... ", end="")
        text, ref = dic["text"], dic["tokens"]
        tokens = tk_function(text)
        if ref == tokens:
            print("ok!")
        else:
            errors += 1
            missing = set(ref).difference(tokens)
            leftover = set(tokens).difference(ref)
            print(f"errors: {list(leftover)} vs {list(missing)}")
    print("-"*30)
    if errors == 0:
        print("Good work!")
    else:
        print(f"{errors} errors, keep working!")


# testing
filename = "input-tokenizer.json"
test_tokenizer(tokenize, filename)
```
